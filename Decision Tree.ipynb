```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
concepts=np.array(data)[:,:-1]
concepts
```

```python
target=np.array(data)[:,-1]
target
```

```python
def train(con,tar):
  for i,val in enumerate(tar):
    if val=="yes":
      specific_h=con[i].copy()

  for i,val in enumerate(con):
    if tar[i]=="yes":
      for x in range (len(specific_h)):
        if val[x] != specific_h[x]:
          specific_h[x]="?"
        else:
          pass
    return specific_h

# here con is the concept and tar is the taget
#copy is used to create a reference
```

```python
print(train(concepts,target))
```
