```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
data.plot(kind='scatter',x='Age',y='Educ')
plt.show()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data.Age,data.Age)
```

```python
plt.scatter(X_train,Y_train,label="Traning data set",color='r',alpha=0.7)
plt.scatter(X_test,Y_test,label="Testing data set",color='g',alpha=0.7)
plt.legend()
plt.title("Test the Train Split")
plt.show()
```

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1),Y_train.values)
```

```python
prediction = LR.predict(X_test.values.reshape(-1,1))

plt.plot(X_test,prediction,label="Linear Regression",color='r',alpha=0.7)
plt.scatter(X_test,Y_test,label="Actual Test Data set",color='g',alpha=0.7)
plt.legend()
```
